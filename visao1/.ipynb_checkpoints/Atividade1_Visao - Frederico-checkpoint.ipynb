## Atividade 1

Realize as seguintes atividades:

1. O histograma fornecido aproveita bem as faixas de cores?
2. Equalize o histograma de maneira que a cumulativa seja uma função linear na diagonal do gráfico
3. Plote a imagem desenhada

**1.1** Não. o histograma fornecido só apresentava cores na escala de cinza até a intensidade 58, sendo que aceita até 255. Logo, o espectro estava "comprimido" em uma pequena faixa.

**1.2**

```python
rintin_eqlz = rintin_grey

m = (255.0)/(rintin.max() - rintin.min())
h = (255/(rintin.max() - rintin.min()))*rintin.min()

for i in range(rintin_grey.shape[0]):
    for j in range(rintin_grey.shape[1]):
        rintin_eqlz[i][j] =  m * (rintin_grey[i][j] - rintin_grey.min())
```

```python
cv2.imwrite("equalizado.png",rintin_eqlz)
```

```python
print(m)
```

```python
hist,bins = np.histogram(rintin_eqlz.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()


# cdf_m = np.ma.masked_equal(cdf,0)
# cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
# cdf = np.ma.filled(cdf_m,0).astype('uint8')
# img2 = cdf[rintin_grey]
# hist2,bins2 = np.histogram(img2.flatten(),256,[0,256])


plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_eqlz.flatten(),256,[0,256], color = 'r')
plt.xlim([rintin_eqlz.flatten().min(),256])
plt.legend(('cdf','histogram'), loc = 'upper right')
plt.show()

```
